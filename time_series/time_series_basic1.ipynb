
# Data Source 

- Use Pandas Data Source `pandas_datareader` to access real financial data 

## Note 

- Check which of the supported Data Source is actually available

```
import pandas_datareader.data as web
import datetime
```

```
start = datetime.datetime(2018, 1, 1)
end = datetime.datetime(2018, 12, 1)
f = web.DataReader('AAPL', 'iex', start, end)
f.head()
```


# Plot
