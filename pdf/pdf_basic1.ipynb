
# Overview 

The fundamental element of PDF is can be represented in 2 ways in the Python environment 

## 1. Numpy Vector 

- It can be implemented as a pair of numpy vectors `x` and `y`and the specific `i` position is used to associate a certain $x$ in the Input Domain to the corresponding $y$ Probability  

## 2. Callable Object 

- This is the closer representation to the $f(x)$ mathematical definition and it can be implemented as a Callable `f(v)` with `v` a Numpy Array containing the domain values 

- This kind of representation has the significant advantage over the vector representation to be more generic as it can get evaluated on a lazy way in different times and in different domain regions 



```
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```


Let's define the domain of interest 


```
x = np.arange(-10, 10, 0.01)
```


Let's represent the PDF according to definition 1 


```
# Example of Direct Definition 
y1 = stats.norm.pdf(x,0,2)
```


Let's plot specifying both domain and codomain 


```
# NOTE: The domain and codomain association is performed correctly as both x and y are passed to plot 
plt.plot(x, y1)
```


If domain is omitted, a default is used and the resulting graph is not correct 


```
# NOTE: Try to see what happens if you do not pass x --> The domain is wrong as expected 
plt.plot(y1)
```


Let's represent the PDF according to definition 2 


```
# We need functools to define callable objects as a result of partial arguments application 
import functools
```


Let's define the callable 


```
# NOTE: The generic PDF is defined via the Arguments Partial Application 
f = functools.partial(stats.norm.pdf, loc=0, scale=2) 
```


Let's evaluate the callable 


```
y2 = f(x)
```


Let's plot


```
plt.plot(x, y2)
```
